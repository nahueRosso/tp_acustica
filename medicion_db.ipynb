# Medición En Tiempo real de db

### 1- ejecutar el codigo para obtener la presion sonora

```python
import sounddevice as sd
import numpy as np

# Función de callback que se llama en cada bloque de audio
def audio_callback(indata, frames, time, status):
    if status:
        print(f"Error en la captura de audio: {status}")
    if any(indata):
        # Calcula el nivel de decibeles en tiempo real
        rms = np.sqrt(np.mean(indata**2))
        decibel = 20 * np.log10(rms)
        print(f"Nivel de decibeles: {decibel:.2f} dB")

# Configuración de la captura de audio
sample_rate = 44100  # Tasa de muestreo (puedes ajustarla según tus necesidades)
block_size = 1024   # Tamaño del bloque de audio

with sd.InputStream(callback=audio_callback, channels=1, samplerate=sample_rate, blocksize=block_size):
    print("Medición de decibeles en tiempo real. Presiona Ctrl+C para detener.")
    try:
        while True:
            pass
    except KeyboardInterrupt:
        pass
```
