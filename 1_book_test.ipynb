- Importe algunas librerias 

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
```

### Use el modulo librosa para manejar archivos de audio

```python
audio_path = './impulse_responses_db/2200BC model/s1-m1/IR_4000.wav'
```

```python
wave_type, sample_rate = librosa.load(audio_path)
spectrogram = librosa.feature.melspectrogram(wave_type, sr=sample_rate)
log_spectrogram = librosa.power_to_db(spectrogram, ref=np.max)
```

- Analise los datos del audio

```python
print(f'sample_rate {(sample_rate)}')
print(f'wave_type  tipo : {type(wave_type)}, largo : {len(wave_type)}')
print(f'spectrogram  tipo : {type(spectrogram)}, largo : {len(spectrogram)}')
print(f'log_spectrogram  tipo : {type(log_spectrogram)}, largo : {len(log_spectrogram)}')
```

### Hice un grafico de espectro

```python
plt.figure(figsize=(10, 6))
plt.imshow(log_spectrogram, aspect='auto', origin='lower', cmap='hot')
plt.colorbar(format='%+2.0f dB')
plt.title('Espectrograma')
plt.show()
```
