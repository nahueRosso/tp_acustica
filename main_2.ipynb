```python
import numpy as np
import matplotlib.pyplot as plt
import librosa

# Carga el archivo WAV
audio_path = './impulse_responses_db/2200BC model/s5-m1/IR_4000.wav'
y, sr = librosa.load(audio_path, sr=None)

# Normaliza el audio a 0 dB
max_amplitude = np.max(np.abs(y))
y_normalized = y / max_amplitude

# Calcula el tiempo en segundos para el eje x
time = np.arange(0, len(y)) / sr

# Convierte la amplitud a dB (decibelios)
y_dB = 20 * np.log10(np.abs(y_normalized))

# Grafica el audio normalizado en dB
plt.figure(figsize=(10, 6))
plt.plot(time, y_dB)
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud (dB)')
plt.title('Audio Normalizado en 0 dB')
plt.grid(True)
plt.show()
```

```python
# Encuentra el nivel de inicio y el nivel de fin
nivel_inicio = y_dB[0]  # Puedes ajustar este valor según lo que veas en tu gráfico
nivel_fin = nivel_inicio - 60  # Nivel de fin, 60 dB por debajo del nivel de inicio

# Encuentra el tiempo en el que cae desde el nivel de inicio al nivel de fin
for i, amplitud_dB in enumerate(y_dB):
    if amplitud_dB <= nivel_fin:
        tiempo_RT60 = time[i]
        break

# Imprime el valor de RT60
print(f"RT60: {tiempo_RT60:.2f} segundos")
```

```python
import numpy as np
from scipy.optimize import curve_fit
import librosa

# Carga la señal de audio
audio_file = './impulse_responses_db/2200BC model/s5-m1/IR_4000.wav'
y, sr = librosa.load(audio_file)

# Aísla la cola de reverberación (puedes necesitar ajustar estos valores)
start_sample = int(sr * 0.1)  # Ejemplo: 100 ms de inicio
end_sample = -int(sr * 0.5)    # Ejemplo: 500 ms de final

reverb_tail = y[start_sample:end_sample]

# Calcula la energía de la señal en función del tiempo
energy = np.cumsum(reverb_tail**2)

# Define una función exponencial para el ajuste de curvas
def exponential_decay(t, A, tau):
    return A * np.exp(-t / tau)

# Estima los parámetros iniciales para el ajuste
max_energy = np.max(energy)
initial_A = max_energy
initial_tau = len(energy) / (4 * sr)  # Aproximadamente un cuarto de segundo

# Ajusta la curva exponencial a la curva de energía
time = np.arange(len(energy)) / sr
params, _ = curve_fit(exponential_decay, time, energy, p0=(initial_A, initial_tau))

# Calcula el RT60 como 5.0 * tau (para una disminución del 99%)
rt60 = 5.0 * params[1]

print(f"RT60 estimado: {rt60} segundos")
```
